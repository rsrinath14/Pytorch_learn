```
import pandas as pd
data = pd.read_csv("https://raw.githubusercontent.com/PacktWorkshops/The-Deep-Learning-with-PyTorch-Workshop/master/Chapter02/Activity2.01/YearPredictionMSD.csv",nrows = 50000)
data.head()
```

```
data.dtypes
```

```
data.isnull().sum().sum()
```

```
outliers = {}
for i in range(data.shape[1]):
    min_t = data[data.columns[i]].mean() - (3 * data[data.columns[i]].std())
    max_t = data[data.columns[i]].mean() + (3 * data[data.columns[i]].std())
    count = 0
    for j in data[data.columns[i]]:
        if j < min_t or j > max_t:
            count += 1
    percentage = count/data.shape[0]
    outliers[data.columns[i]] = "%.3f" % percentage
print(outliers)
```

```
X = data.iloc[:, 1:]
Y = data.iloc[:, 0]
```

```
X = (X - X.mean())/X.std()
X.head()
```

```
from sklearn.model_selection import train_test_split
X_shuffle = X.sample(frac=1, random_state=0)
Y_shuffle = Y.sample(frac=1, random_state=0)
x_new, x_test, y_new, y_test = train_test_split(X_shuffle, 
                                                Y_shuffle, 
                                                test_size=0.2, 
                                                random_state=0)
dev_per = x_test.shape[0]/x_new.shape[0]
x_train, x_dev, y_train, y_dev = train_test_split(x_new, 
                                                  y_new, 
                                                  test_size=dev_per, 
                                                  random_state=0)
```
